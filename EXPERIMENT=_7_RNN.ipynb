# **Implement RNN using TensorFlow**

**Theory**

A Recurrent Neural Network (RNN) is a neural network designed to process sequential data. Unlike feed-forward networks, RNNs maintain a hidden state that carries information from previous time steps.

The hidden state is calculated as:

image.png

image.png

```python
import tensorflow as tf
import numpy as np
```

```python
# Input sequence
x = [1, 2, 3]

# Given weights
Wh = 0.5
Vh = 0.3
Wy = 0.7

# Biases
bh = 0.1
by = 0.2

# Initial hidden state
h_prev = 0
```

**RNN FORWARD PASS**

```python
for i in range(len(x)):

    xt = x[i]

    # Hidden state calculation
    h = tf.math.tanh(Wh * xt + Vh * h_prev + bh)

    # Output calculation
    y = Wy * h + by

    print("Time Step:", i+1)
    print("Hidden State:", h.numpy())
    print("Output:", y.numpy())
    print()

    # Update hidden state
    h_prev = h
```

For every time step:

Hidden state is computed using tanh activation.

Output is computed using the hidden state.

Hidden state is passed to the next time step.

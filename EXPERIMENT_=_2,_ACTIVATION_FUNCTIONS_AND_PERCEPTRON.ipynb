# **A)To implement different activation functions commonly used in neural networks using python**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
inputs = np.array([-3, -1, 0, 1, 3])

print("Input values:", inputs)
```



---



**Binary Step Activation Function (Simplest):**

It depends on a threshold value that decides whether a neuron should be activated or not.

This makes it suitable for basic **binary classification tasks**.

**Limitations:**

**Non-differentiability:** The binary step function has a gradient of zero everywhere except at the threshold. This makes it unsuitable for backpropagation, as the gradient-based optimization algorithms cannot update weights effectively.

**Inability to handle complex patterns:** Being a piecewise linear function, it lacks the non-linearity required to model complex data representations.

**Limited to binary outputs:** It cannot handle multi-class problems or provide probabilistic outputs.

**Mathematical formula:**

image.png

```python
def binary_step(x):
    if x >= 0:
        return 1
    else:
        return 0

output = [binary_step(i) for i in inputs]

print("Binary Step Output:", output)
```

*Expected behavior: negative → 0, positive → 1.*


---



**Linear Activation Function**

**Formula:**

image.png

```python
def linear(x):
    return x

output = [linear(i) for i in inputs]

print("Linear Output:", output)
```

```python
print("Linear Output:", np.array(output))
```

**Why is linear activation rarely used in hidden layers?**

1) Because it does not introduce non-linearity

2) A neural network with only linear activations behaves like a single linear model, no matter how many layers it has.


---



**Sigmoid Activation Function**

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

output = [sigmoid(i) for i in inputs]

print("Sigmoid Output:", output)
```

```python
print("Sigmoid activation output:", np.array(output))
```



---



**Tanh Activation Function**

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

output = [tanh(i) for i in inputs]

print("Tanh Output:", output)
```



---



**ReLU Activation Function**

```python
def relu(x):
    if x > 0:
        return x
    else:
        return 0

output = [relu(i) for i in inputs]

print("ReLU Output:", output)
```

Negative values become 0.


---



**Leaky ReLU Activation Function**

```python
def leaky_relu(x):
    if x > 0:
        return x
    else:
        return 0.01 * x

output = [leaky_relu(i) for i in inputs]

print("Leaky ReLU Output:", output)
```

Prevents dying ReLU problem.


---



**Softmax Activation Function**

Softmax works on a vector of values, not individual numbers.

```python
def softmax(x):
    exp_values = np.exp(x)
    return exp_values / np.sum(exp_values)

sample_input = np.array([2.0, 1.0, 0.1])

output = softmax(sample_input)

print("Input:", sample_input)
print("Softmax Output:", output)
print("Sum of probabilities:", np.sum(output))
```

Softmax outputs probabilities that sum to 1.

**VIZUALIZATIONS OF THESE ACTIVATION FUNCTIONS:**

```python
x = np.linspace(-10, 10, 100)
```

```python
def binary_step(x):
    return np.where(x >= 0, 1, 0)

def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x):
    return np.where(x > 0, x, 0.01 * x)
```

```python
y_step = binary_step(x)
y_linear = linear(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky = leaky_relu(x)
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.plot(x, y_step)
plt.title("Binary Step")

plt.subplot(2,3,2)
plt.plot(x, y_linear)
plt.title("Linear")

plt.subplot(2,3,3)
plt.plot(x, y_sigmoid)
plt.title("Sigmoid")

plt.subplot(2,3,4)
plt.plot(x, y_tanh)
plt.title("Tanh")

plt.subplot(2,3,5)
plt.plot(x, y_relu)
plt.title("ReLU")

plt.subplot(2,3,6)
plt.plot(x, y_leaky)
plt.title("Leaky ReLU")

plt.tight_layout()
plt.show()
```

```python
sample_input = np.array([2.0, 1.0, 0.1])
softmax_output = softmax(sample_input)

print("Input:", sample_input)
print("Softmax Output:", softmax_output)
print("Sum of probabilities:", np.sum(softmax_output))
```

```python
plt.subplot(3,3,7)
plt.bar(range(len(softmax_output)), softmax_output)
plt.title("Softmax")
plt.xlabel("Class Index")
plt.ylabel("Probability")
```

# **B)** **Classification Using Perceptron**

The formula used in a perceptron is:

image.png

image.png

```python
import numpy as np
```

```python
# Data points
X = np.array([
    [3, 2],
    [0, -1],
    [-2, 1]
])

# Weights
w1 = -0.7
w2 = 0.3

# Bias
b = -0.2

print("Input Data:\n", X)
print("Weights:", w1, w2)
print("Bias:", b)
```

```python
z_values = []

for x1, x2 in X:
    z = (w1 * x1) + (w2 * x2) + b
    z_values.append(z)

print("Net Input (z) values:", z_values)
```

```python
def step_function(z):
    if z >= 0:
        return 1
    else:
        return 0
```

```python
step_outputs = [step_function(z) for z in z_values]

print("Step Activation Outputs:", step_outputs)
```

**Explanation:**

First two inputs give negative z → class 0

Third input gives positive z → class 1

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
sigmoid_outputs = [sigmoid(z) for z in z_values]

print("Sigmoid Outputs:", sigmoid_outputs)
```

```python
print("\nFinal Results:")

for i in range(len(X)):
    print(f"Input: {X[i]}, z = {z_values[i]:.2f}, Step = {step_outputs[i]}, Sigmoid = {sigmoid_outputs[i]:.3f}")
```

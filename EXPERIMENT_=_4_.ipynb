# **4.	To Implement AND, OR, NOR, NAND, XOR, and XNOR using ANN**

**AND Gate using ANN**

Explanation: Output is 1 only when both inputs are 1.

```python
import numpy as np

def step(x):
    return 1 if x >= 0 else 0

# weights and bias for AND gate
w = np.array([1, 1])
b = -1.5

inputs = np.array([[0,0],[0,1],[1,0],[1,1]])

print("AND Gate Output:")
for x in inputs:
    y = step(np.dot(w,x) + b)
    print(x,"->",y)
```

**OR Gate using ANN**

Explanation: Output is 1 if at least one input is 1.

```python
import numpy as np

def step(x):
    return 1 if x >= 0 else 0

# weights and bias for OR gate
w = np.array([1,1])
b = -0.5

inputs = np.array([[0,0],[0,1],[1,0],[1,1]])

print("OR Gate Output:")
for x in inputs:
    y = step(np.dot(w,x) + b)
    print(x,"->",y)
```

**NAND Gate using ANN**

Explanation: NAND is the inverse of AND.

```python
import numpy as np

def step(x):
    return 1 if x >= 0 else 0

# weights and bias for NAND gate
w = np.array([-2,-2])
b = 3

inputs = np.array([[0,0],[0,1],[1,0],[1,1]])

print("NAND Gate Output:")
for x in inputs:
    y = step(np.dot(w,x) + b)
    print(x,"->",y)
```

**NOR Gate using ANN**

Explanation: NOR outputs 1 only when both inputs are 0.

```python
import numpy as np

def step(x):
    return 1 if x >= 0 else 0

# weights and bias for NOR gate
w = np.array([-1,-1])
b = 0.5

inputs = np.array([[0,0],[0,1],[1,0],[1,1]])

print("NOR Gate Output:")
for x in inputs:
    y = step(np.dot(w,x) + b)
    print(x,"->",y)
```

**XOR Gate using ANN**

Explanation: XOR is not linearly separable, so we need a multi-layer neural network.

```python
import numpy as np

def step(x):
    return 1 if x >= 0 else 0

def xor_gate(x1,x2):

    # hidden layer
    h1 = step(x1 + x2 - 0.5)   # OR neuron
    h2 = step(x1 + x2 - 1.5)   # AND neuron

    # output layer
    y = step(h1 - h2 - 0.5)

    return y

inputs = [(0,0),(0,1),(1,0),(1,1)]

print("XOR Gate Output:")
for x in inputs:
    print(x,"->",xor_gate(x[0],x[1]))
```

**XNOR Gate using ANN**

Explanation: XNOR is the inverse of XOR.

```python
import numpy as np

def step(x):
    return 1 if x >= 0 else 0

def xnor_gate(x1,x2):

    # hidden layer
    h1 = step(x1 + x2 - 0.5)
    h2 = step(x1 + x2 - 1.5)

    xor = step(h1 - h2 - 0.5)

    # invert XOR
    return 1 - xor

inputs = [(0,0),(0,1),(1,0),(1,1)]

print("XNOR Gate Output:")
for x in inputs:
    print(x,"->",xnor_gate(x[0],x[1]))
```

**Key Concept:**

Single layer perceptron works for: AND, OR, NAND, NOR

Multi-layer ANN required for: XOR, XNOR (because they are not linearly separable)

**THEORY** = This experiment aims to demonstrate the fundamental operations Python performs on numeric (integers, floats) and alphabetic/string data types. Python's dynamism allows it to select appropriate operations based on the operand type.For numeric data, the standard arithmetic operators ( + ,  − ,  ∗ ,  / ,  % ,  ∗∗ ) are executed. For alphabetic (string) data, the meaning of these operators changes: addition ( + ) performs concatenation (joining strings), and multiplication ( ∗ ) performs repetition (duplicating the string). This fundamental difference in operator behavior, or operator overloading, is key to performing versatile data manipulations efficiently in Python.
